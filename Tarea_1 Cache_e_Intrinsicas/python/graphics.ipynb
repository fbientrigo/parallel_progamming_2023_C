```python
import numpy as np
import matplotlib.pyplot as plt

N = 2048

# en base al pdf Respuesta1.pdf
def T(N,p):
    return (5 + 1/2**10)*np.log2(p) + (N/p) * (  (p-1)/2**10 + 1 ) - 1
```

```python
T1 = T(N,1)
print(f"Para un solo procesador {T1} [ms]")
```

```python
p_arr = []
T_arr = []
S_arr = []
E_arr = []

for PE in [2**i for i in range(1,11)]:
    print(f"Para {PE} unidades de procesamiento")
    Tp = T(N, PE)
    Sp = T1/Tp
    Ep = Sp/PE
    print(f"    Demora: {Tp} [ms]")
    print(f"    Speedup: {Sp}")
    print(f"    Eficiencia: {Ep}")

    p_arr.append(PE)
    T_arr.append(Tp)
    S_arr.append(Sp)
    E_arr.append(Ep)
```

```python
fig, ax1 = plt.subplots() 

ax1.plot(p_arr, T_arr,
         label="Tiempo [ms]")
ax1.scatter(p_arr, T_arr,
         label="Tiempo [ms]")
ax1.set_xlabel("Procesadores p")
ax1.set_ylabel('Tiempo [ms]', color = 'blue') 
ax1.tick_params(axis ='y', labelcolor = 'blue') 

ax2 = ax1.twinx()
ax2.set_ylabel('Speedup', color = 'orange') 
ax2.scatter(p_arr, S_arr,
         label="Speedup", color="orange")
ax2.plot(p_arr, S_arr,
         label="Speedup", color="orange")
ax2.tick_params(axis ='y', labelcolor = 'orange') 

plt.title("Analisis de Tiempo y SpeedUp")
plt.grid()
plt.show()
```

```python
fig, ax1 = plt.subplots() 

ax1.plot(p_arr, E_arr,
         label="Eficiencia")
ax1.scatter(p_arr, E_arr,
         label="Eficiencia")
ax1.set_xlabel("Procesadores p")
ax1.set_ylabel('Eficiencia', color = 'blue') 
ax1.tick_params(axis ='y', labelcolor = 'blue') 

ax2 = ax1.twinx()
ax2.set_ylabel('Speedup', color = 'orange') 
ax2.scatter(p_arr, S_arr,
         label="Speedup", color="orange")
ax2.plot(p_arr, S_arr,
         label="Speedup", color="orange")
ax2.tick_params(axis ='y', labelcolor = 'orange') 

plt.title("Analisis de SpeedUp y Eficiencia")
plt.grid()
plt.show()
```
